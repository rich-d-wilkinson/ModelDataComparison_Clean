The problem in a nutshell is that if we thin the grid, then the ordering of the models according to their loglikelihood score varies considerably.



```python
# @hidden_cell
import GPy
import sys
import os
sys.path.append(os.getenv("HOME") + "/Documents/Code/Emulation/GPyDifferentMetrics/")
from HaversineDist import Exponentialhaversine
import numpy as np
%matplotlib inline
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import matplotlib
import glob as glob
from cmocean import cm as cm
from Utilities import *


GCM_dir = os.getenv("HOME")+'/Documents/Code/ModelDataComparison/DMC/Scripts/Model_data/CO2_anom/'
gcm_SSTs = glob.glob(GCM_dir+'t*.txt')

gcm_mask = np.genfromtxt(GCM_dir+'mask.txt', dtype='int')

obs_dir = os.getenv("HOME")+'/Documents/Code/ModelDataComparison/DMC/Scripts/Observation_data/P3+_SST_anom/'
file = 'lambda_10.txt'
observations = np.genfromtxt(obs_dir+file, skip_header=1)

X_obs = observations[:,0:2]
y_obs = observations[:,2].reshape(-1,1)
var_ratios = observations[:,3][:,None]**2

from scaledheteroscedasticgaussian import ScaledHeteroscedasticGaussian
from gp_heteroscedastic_ratios import ScaledHeteroscedasticRegression

k3 = Exponentialhaversine(2, lengthscale=2000)
m3 = ScaledHeteroscedasticRegression(X=X_obs, Y=y_obs, kernel=k3, noise_mult=1., 
                                     known_variances=var_ratios)
m3.optimize_restarts(10)

print(m3)
```

```python
count=0
gcm_runs = np.zeros((8,27186))
gcm_runs_label = gcm_SSTs.copy()

for file_name in gcm_SSTs:
    file_nm = file_name.split(GCM_dir)[-1]
    #print(file_nm)
    # Read in GCM output.
    gcm_runs[count,:] = np.genfromtxt(file_name)
    gcm_runs_label[count] = file_nm.split(".txt")[0]
    count +=1

all_gcm_thinned = {}
all_X_pred = {}
for thinby in (np.arange(6)+2):
    tmp1, tmp2 = ThinGrid(gcm_runs[0,:], gcm_mask, thinby=thinby) # just to get size
    gcm_thin = np.zeros((8, tmp2.size))
    for count in range(8):
        X_pred, out = ThinGrid(gcm_runs[count,:], gcm_mask, thinby=thinby)
        gcm_thin[count,:] = out.flatten()
        
    all_gcm_thinned[str(thinby)]= gcm_thin
    all_X_pred[str(thinby)]=X_pred 
    
```

```python

Chols = {}
mus = {}
Covs = {}
for thinby in (np.arange(6)+2):
    mus[str(thinby)], Covs[str(thinby)] = m3.predict_noiseless(all_X_pred[str(thinby)], full_cov=True)
    Chols[str(thinby)] = np.linalg.cholesky(Covs[str(thinby)])
    
```

```python
from Cholesky import *
loglikes={}
orderings={}
relative={}

for thinby in (np.arange(6)+2):
    loglikes[str(thinby)]=dlogmvnorm(all_gcm_thinned[str(thinby)].T, mus[str(thinby)], Chols[str(thinby)])
    orderings[str(thinby)] = np.argsort(-loglikes[str(thinby)]) #minus sign so that max is first
    relative[str(thinby)] = np.round(loglikes[str(thinby)]-np.max(loglikes[str(thinby)]),1)
```

If we look at the ordering of the model in terms of the log-likelihood scores, then they vary according to how much the grid is thinned. Here, 2 denotes that every second grid cell of the GCM output (both in lat and long) was used to evaluate the score (thinning to every second value gives the same answer as using the full grid - which I've not done here as it takes 10 minutes on my computer every time I do it). 3 denotes that we took every third value etc.

The numbers in the array are the order of the models from best to worse.

So we can see that we move from favouring low values of CO2 when we use a fine grid (e.g, thinby=2), to favouring values closer to 400ppm if we use a coarse grid.

Let's compare this to the RMSE, and check whether this changes for different grid resolutions.

```python
def MSE(X_obs, X_pred, y_obs, gcm_out):
    k = Exponentialhaversine(2)
    index = np.argmin(k._unscaled_dist(X_obs, X_pred), axis=1)
    #print(k._unscaled_dist(X_obs, X_pred)[range(y_obs.size),index])
    y_gcm_grid = gcm_out[index]
    return(np.sqrt(np.mean((y_gcm_grid-y_obs)**2)))    
X_pred, out = ThinGrid(gcm_runs[0,:], gcm_mask, thinby=1)


RMSEs={}
orderingsRMSEs={}



for thinby in (np.arange(6)+2):
    MSEs = np.zeros(8)
    for ii in range(8):
        MSEs[ii]=(MSE(X_obs, all_X_pred[str(thinby)], y_obs, all_gcm_thinned[str(thinby)][ii,:]))
    RMSEs[str(thinby)]=MSEs
    orderingsRMSEs[str(thinby)] = np.argsort(MSEs)
    #print(orderingsRMSEs)
```

```python
print(orderingsRMSEs)
ind = np.argsort(CO2)


for thinby in (np.arange(6)+2):
    plt.scatter(CO2, -RMSEs[str(thinby)], color=cols[thinby-2])
    plt.plot(CO2[ind], -RMSEs[str(thinby)][ind],color=cols[thinby-2])
    plt.legend()

plt.title('RMSE vs pCO2')
```

So thinning also changes the order of the RMSEs, but not to a worrying degree.


What seems to be happening is that the short range correlations are unduly influencing the log-likelihood. As we use a coarser grid, the short range correlations aren't included in the calculation, and so the ordering changes.

The exponential kernel is a Markovian kernel, we can do the same thing using a non-Markovian kernel such as the RBF, and check with this still happens.

```python
def Haversine(k):
    from functools import partial
    k._unscaled_dist = partial(Exponentialhaversine._unscaled_dist, k)
    return k

kRBF = Haversine(GPy.kern.RBF(2, lengthscale=4000))
mRBF = ScaledHeteroscedasticRegression(X=X_obs, Y=y_obs, kernel=kRBF, noise_mult=1., 
                                     known_variances=var_ratios)
mRBF.optimize_restarts(10)

```

```python
CholsRBF = {}
musRBF = {}
CovsRBF = {}
for thinby in (np.arange(6)+2):
    musRBF[str(thinby)], CovsRBF[str(thinby)] = mRBF.predict_noiseless(all_X_pred[str(thinby)], full_cov=True)
    CholsRBF[str(thinby)] = np.linalg.cholesky(CovsRBF[str(thinby)]+np.identity(musRBF[str(thinby)].size)*0.0001)

```

```python
loglikesRBF={}
orderingsRBF={}
relativeRBF={}

for thinby in (np.arange(6)+2):
    loglikesRBF[str(thinby)]=dlogmvnorm(all_gcm_thinned[str(thinby)].T, musRBF[str(thinby)], CholsRBF[str(thinby)])
    orderingsRBF[str(thinby)] = np.argsort(-loglikesRBF[str(thinby)]) #minus sign so that max is first
    relativeRBF[str(thinby)] = np.round(loglikesRBF[str(thinby)]-np.max(loglikesRBF[str(thinby)]),1)

```

```python
orderingsRBF
```

```python
for thinby in (np.arange(6)+2):
    plt.scatter(CO2, -relativeRBF[str(thinby)]/relativeRBF[str(thinby)].min(), color=cols[thinby-2])
    plt.plot(CO2[ind], -relativeRBF[str(thinby)][ind]/relativeRBF[str(thinby)].min(),color=cols[thinby-2])
    plt.legend()

plt.title('Standardized vs pCO2')
```

So the ordering does change slightly, but not by much (apart from that weird effect for the yellow curve).  However, this is not what Fran and Dan were expecting. They believe moderate pCO2 values closer to 400ppm are most sensible. So this result certainly doesn't solve the problem.

### Using just the mean output field

Let's check it is definitely the covariance causing this problem, by just comparing the mean fields with the GCM results.

```python

mean_RMSE = {}
orderings_mean={}
for thinby in (np.arange(6)+2):
    mean_RMSE[str(thinby)]=np.sqrt(np.mean((mus[str(thinby)]-all_gcm_thinned[str(thinby)].T)**2, axis=0))
    orderings_mean[str(thinby)] = np.argsort(mean_RMSE[str(thinby)])
```

```python

print(orderings_mean)
ind = np.argsort(CO2)


for thinby in (np.arange(6)+2):
    plt.scatter(CO2, -mean_RMSE[str(thinby)], color=cols[thinby-2])
    plt.plot(CO2[ind], -mean_RMSE[str(thinby)][ind],color=cols[thinby-2])
    plt.legend()

plt.title('Standardized vs pCO2')
```


These curves show the RMSEs for comparing the mean from the GP (fitted to the data) with the GCM runs, thinned to various degrees. These results are remarkably consistent. 

So it definitely is the covariance causing the problem. 

The next thing to check, is what happens if we ignore the covariance structure, and just use the variance from the GP, but ignore the covariance, and use this to calculate the loglikelihood.

```python
Chols_diag = {}
mus = {}
Covs_diag = {}
for thinby in (np.arange(6)+2):
    mus[str(thinby)], Covs_diag[str(thinby)] = m3.predict_noiseless(all_X_pred[str(thinby)], full_cov=False)
    Chols_diag[str(thinby)] = np.linalg.cholesky(np.diagflat(Covs_diag[str(thinby)]))
```

```python
loglikes_diag={}
orderings_diag={}
relative_diag ={}
for thinby in (np.arange(6)+2):
    loglikes_diag[str(thinby)]=dlogmvnorm(all_gcm_thinned[str(thinby)].T, mus[str(thinby)], Chols_diag[str(thinby)])
    orderings_diag[str(thinby)] = np.argsort(-loglikes_diag[str(thinby)]) #minus sign so that max is first
    relative_diag[str(thinby)] = np.round(loglikes_diag[str(thinby)]-np.max(loglikes_diag[str(thinby)]),1)

```

```python
print(orderings_diag)
ind = np.argsort(CO2)

for thinby in (np.arange(6)+2):
    plt.scatter(CO2, -relative_diag[str(thinby)]/relative_diag[str(thinby)].min(), color=cols[thinby-2])
    plt.plot(CO2[ind], -relative_diag[str(thinby)][ind]/relative_diag[str(thinby)].min(),color=cols[thinby-2])
    plt.legend()


plt.title('Standardized vs pCO2')
```

So again, we now get stable results. So it is definitely the covariance structure causing the problem.
